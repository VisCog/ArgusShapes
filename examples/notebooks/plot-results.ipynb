```python
import numpy as np
import pandas as pd
import pulse2percept as p2p
import p2pspatial
import pickle

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pickle.load(open('../../scripts/crossval-spatial_decay_ahuja_128_2017-11-16_19-57-41.pickle', 'rb'))
```

```python
X_test, y_true, y_pred, cv_results, specifics = data
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(14, 6))
for j, axname in enumerate(zip(axes.ravel(), ['area', 'orientation', 'major', 'minor'])):
    ax, name = axname
    for yt, yp in zip(y_true, y_pred):
        yp = pd.DataFrame(yp)
        if j >= yp.shape[1]:
            continue
        ax.plot(yt.iloc[:, j], yp.iloc[:, j], 'o')
    lims = ax.axis()
    ax.plot([lims[0], lims[1]], [lims[0], lims[1]], 'k--')
    ax.set_xlabel('true ' + name)
    ax.set_ylabel('predicted ' + name)
    ax.set_title(name)
    ax.axis('square')
fig.tight_layout()
```

```python
scores = []
for cvr in cv_results:
    scores.append(cvr['mean_test_score'])
scores = np.array(scores)
```

```python
decay_const = cv_results[0]['param_reg__decay_const'].data
plt.errorbar(decay_const, np.mean(scores, axis=0), yerr=np.std(scores, axis=0))
plt.xlabel('decay constant')
plt.ylabel('R2')
```
