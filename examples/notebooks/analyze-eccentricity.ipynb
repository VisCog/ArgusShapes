```python
import os
import numpy as np
import pandas as pd
import pyswarm
import scipy.stats as sps

import argus_shapes
import pulse2percept.implants as p2pi
from pulse2percept.retina import ret2dva, dva2ret

import sklearn.base as sklb
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-talk')
```

```python
subject_params = {
    '12-005': {
        'implant_type': p2pi.ArgusII,
        'implant_x': -1344.36597,
        'implant_y': 537.7463881,  # or should this be minus?
        'implant_rot': -0.664813628,
        'loc_od_x': 15.5,
        'loc_od_y': 1.2,
        'marker': 'o'
    },
    '51-009': {
        'implant_type': p2pi.ArgusII,
        'implant_x': 398.514982,
        'implant_y': -540.8417613,
        'implant_rot': -0.526951314,
        'loc_od_x': 14.8,
        'loc_od_y': 4.7,
        'marker': 's'
    },
    '52-001': {
        'implant_type': p2pi.ArgusII,
        'implant_x': -1147.132944,
        'implant_y': -369.1922119,
        'implant_rot': -0.342307766,
        'loc_od_x': 14.9,
        'loc_od_y': 4.3,
        'marker': '^'
    }
}
```

```python
subjects = list(subject_params.keys())
subjects
```

```python
y_trial = {}
y_grouped = {}
for subject in subjects:
    implant_setup = subject_params[subject]
    X, y = argus_shapes.load_data(os.path.join(os.environ['SECOND_SIGHT_DATA'], 'shape', subject), 
                                amplitude=2.0, single_stim=True)
    y.loc[:, 'electrode'] = ['%s%d' % (row['electrode'][0], int(row['electrode'][1:])) for _, row in y.iterrows()]
    y_trial[subject] = y
    y_grouped[subject] = y.groupby('electrode', as_index=False).agg(['mean', 'std'])
```

```python
def plot_eccentricity(subject, y, fig=None, axes=None, fit=True):
    implant = p2pi.ArgusII(x_center=implant_setup['implant_x'], 
                           y_center=implant_setup['implant_y'],
                           rot=implant_setup['implant_rot'])
    ecc = [np.sqrt(implant[electrode].x_center ** 2 + implant[electrode].y_center ** 2)
           for electrode, _ in y.iterrows()]
    ecc = np.array(ret2dva(ecc)).reshape(-1, 1)
    params = ['area', 'major_axis_length', 'minor_axis_length', 'orientation']
    if fig is None and axes is None:
        fig, axes = plt.subplots(ncols=len(params), figsize=(14, 3), sharex=True, squeeze=True)
    for param, ax in zip(params, axes):
        mu = np.array(y[param]['mean']).reshape(-1, 1)
        std = np.array(y[param]['std']).reshape(-1, 1)
        if param == 'orientation':
            # Special treatment
            param += ' (deg)'
            mu = np.rad2deg(mu)
            std = np.rad2deg(std)
            ymin = -180
            ymax = 180
            ax.set_yticks(np.linspace(ymin, ymax, 5))
        else:
            ymin = 0
            ymax = np.max(mu)
            scale = 0.5 * 10 ** np.ceil(np.log10(ymax))
            ymax = np.ceil(ymax / scale) * scale
        ax.errorbar(ecc, mu, yerr=std, marker=subject_params[subject]['marker'], 
                    ls='none', label=subject)
        ax.set_xlabel('eccentricity (um)')
        ax.set_ylabel(param)
        if fit:
            linreg = LinearRegression()
            linreg.fit(ecc, mu)
            ax.plot(ecc, linreg.predict(ecc), 'k')
            pearsonr = sps.pearsonr(ecc, mu)
            ax.text(0, ymax, '$r=$%.3f, $p<$%.2e' % (pearsonr[0], pearsonr[1]), 
                    ha='left', va='top', fontsize=10)
        ax.set_xlim(0, 15)
        ax.set_xticks(np.linspace(0, 15, 4))
        ax.set_ylim(ymin, ymax)
    fig.suptitle('Subject ' + subject, fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.9])
```

```python
plot_eccentricity(subject, y_grouped[subject])
```

```python
subject = '51-009'
plot_eccentricity(subject, y_grouped[subject])
```

```python
subject = '52-001'
plot_eccentricity(subject, y_grouped[subject])
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(14, 4), sharex=True, squeeze=True)
for subject in subjects:
    plot_eccentricity(subject, y_grouped[subject], fig=fig, axes=axes, fit=False)
fig.suptitle('')
axes[1].legend()
```
